```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age_group': ['<21', '21-24','24-27','27-30','30-33','33-36','36-39','57-60','>60'],
    'Totle_no_of_loans': [9,310,511,4000,4568,5698,8209,788,6],
    'No_of_badloans': [2,14,20,172,169,188,197,9,0],
    'NO_of_goodloans': [7,296,491,3828,4399,5510,8012,779,6],
    'Percentage_badloans': ['22%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','1.1%','0.0%'],
    'Percentage_goodloans': ['77.8%','95.5%','96.1%','95.7%','96.3%','96.7%','97.6%','98.9%','100%']
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Age_group'], df['Totle_no_of_loans'], color='blue')
plt.xlabel('Age Group')
plt.ylabel('Total Number of Loans')
plt.title('Total Number of Loans by Age Group')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
percent_bad_loans = [float(percent.strip('%')) for percent in df['Percentage_badloans']]
plt.bar(df['Age_group'], percent_bad_loans, color='black')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Bad Loans')
plt.title('Percentage of Bad Loans by Age Group')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
percent_good_loans = [float(percent.strip('%')) for percent in df['Percentage_goodloans']]
plt.bar(df['Age_group'], percent_good_loans, color='blue')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Good Loans')
plt.title('Percentage of Good Loans by Age Group')
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
categories = ['January', 'February', 'March', 'April', 'May']
good_loans = [10000, 15000, 12000, 18000, 20000]
bad_loans = [2000, 2500, 3000, 3500, 4000]
```

```python
plt.figure(figsize=(10, 6))

plt.bar(categories, good_loans, label='Good Loans', color='lightblue')
plt.bar(categories, bad_loans, label='Bad Loans', color='salmon', bottom=good_loans)


plt.xlabel('Month')
plt.ylabel('Loan Amount ($)')
plt.title('Good Loans vs. Bad Loans')
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
