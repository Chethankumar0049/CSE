```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-08-01"]),
    "value":[10,15,11,17,12,18,13]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel('Date')
plt.ylabel("Value")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


date_range = pd.date_range('2022-01-01', '2022-05-01', freq='D')
values = np.random.randn(len(date_range)).cumsum()
data = {'date_column': date_range, 'value_column': values}
df = pd.DataFrame(data)


plt.plot(df['date_column'], df['value_column'], linestyle = 'dotted')


plt.title('Trend and seasonality')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
