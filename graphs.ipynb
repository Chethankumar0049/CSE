```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["r","y","b","g","b"]
sizes=[100,200,300,40,15]
plt.scatter(x,y,c=colors,s=sizes,alpha=1,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```

```python
x1=[3,4,5,6,7]
plt.hist(x1,color="b",edgecolor="g",bottom=1,histtype="bar",orientation="vertical",label="python")
plt.axvline(5,color="g")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
